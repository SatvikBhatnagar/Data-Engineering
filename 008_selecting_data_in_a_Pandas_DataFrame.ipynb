Selecting Data in a Pandas DataFrame

```python
import pandas as pd
file_path = 'data/business-employment-data-march-2023-quarter-industry-revisions.csv'
df = pd.read_csv(file_path)
df.head(2)
```

Boolean Mask

```python
mask = [False for _ in range(len(df))]
mask[3:7] = [True] * 4
mask
```

```python
df[mask] #this returns only the rows that has a True value
#df.loc[mask] this could also work
```

Creating masks using comparison operators

```python
#mask = df.loc[:, 'filled jobs'] < df.loc[:, 'filled jobs revised']
mask = df.loc[:, 'earnings diff'] > 20
mask
```

```python
df[mask]
```

Pandas Boolean Operators

- And:  &
- Or:   |
- Not:  ~
